```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return x ** 2 + 5
```

```python
x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)
y1 = fun(x1)
y2 = fun(x2)
y3 = fun(x3)
plt.plot(x1, y1, label='x^2 + 5')
plt.title('Wykres dla przedziału (-1, 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x2, y2, label='x^2 + 5')
plt.title('Wykres dla przedziału (-6, 6)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x3, y3, label='x^2 + 5')
plt.title('Wykres dla przedziału (0, 5)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
