```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('../Original_Data/10_POMORSKIE.csv')
df_tidy = df.drop(['Unnamed: 0'], axis=1)
df_tidy.to_csv('../Analysis_Data/POMORSKIE.csv', index=False)
df_new = pd.read_csv('../Analysis_Data/POMORSKIE.csv')
```

```python
plt.hist(df_new['Wiek kupującego'], bins=[int(i) for i in range(int(min(df_new['Wiek kupującego'])), int(max(df_new['Wiek kupującego']) + 1))])
plt.title('Rozkład wieku klientów')
plt.xlabel('Wiek')
plt.ylabel('Ilość klientów w danym wieku')
plt.show()
```

```python
oceny = df_new.groupby('Marka')['Ocena'].mean()
fig, ax = plt.subplots()
bar_mean = ax.bar(oceny.index, oceny)
ax.bar_label(bar_mean)
ax.set_title('Wykres średniej oceny dla każdej marki')
ax.set_ylabel('Wartość średniej')
plt.show()
```

```python
marka = df_new.groupby('Marka')['Ocena'].count()
fig, ax = plt.subplots()
bar_count = ax.bar(marka.index, marka)
ax.bar_label(bar_count)
ax.set_title('Wykres ilości zakupionych produktów każdej marki')
ax.set_ylabel('Ilość')
plt.show()
```

```python
sex = df_new.groupby('Płeć kupującego')['Ocena'].count()
fig, ax = plt.subplots()
bar_sex = ax.bar(sex.index, sex)
ax.bar_label(bar_sex)
ax.set_title('Płeć klientów')
ax.set_ylabel('Ilość osób o danej płci')
plt.show()
```

```python
min_day = df_new.groupby('Marka')['Dni od zakupu'].min()
max_day = df_new.groupby('Marka')['Dni od zakupu'].max()
labels = min_day.index
x = np.arange(len(labels))  
width = 0.35  
fig, ax = plt.subplots()
bar_min = ax.bar(x - width / 2, min_day, width, label='Min')
bar_max = ax.bar(x + width / 2, max_day, width, label='Max')
ax.bar_label(bar_min)
ax.bar_label(bar_max)
ax.set_title('Minimalna i maksymalna ilość dni od zakupu')
ax.set_ylabel('Ilość dni')
ax.set_xticks(x, labels)
ax.legend()
plt.show()
```
